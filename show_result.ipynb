```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
data = np.load("../record/outputs.npz")

data.files
```

```python
val_outputs = data['val_out']
val_ins = data['val_in']


for j in range(5):
    tmp_in = val_ins[j*(-1)-1,0]
    plt.figure()
    plt.imshow(tmp_in.reshape(64,-1))
    tmp_out = val_outputs[j,0,:]
    plt.figure()
    plt.imshow(tmp_out.reshape(64,-1))
    
```

```python
test_outputs = data['test_out']
test_ins = data['test_in']
for j in range(10):
    tmp_in = test_ins[j*(-1)-1,0]
    plt.figure()
    plt.imshow(tmp_in.reshape(64,-1))
    tmp_out = test_outputs[j,0,:]
    plt.figure()
    plt.imshow(tmp_out.reshape(64,-1))
```
