```python
import numpy as np
import matplotlib.pyplot as plt
from tools import denormdata
%matplotlib inline

# set display defaults
plt.rcParams['figure.figsize'] = (50, 50)        # large images
plt.rcParams['image.interpolation'] = 'nearest'  # don't interpolate: show square pixels
plt.rcParams['image.cmap'] = 'gray'  # use grayscale output rather than a (potentially misleading) color heatmap
```

```python
#path for records of different units
rec500="../unlimitedrecords/record500-300/outputs.npz"
rec1500="../unlimitedrecords/record1500-300/outputs.npz"
rec2500="../unlimitedrecords/record2500-300/outputs.npz"
```

```python
#get the ddta of inputs and outputs
inputs = np.load(rec500)['test_in']
outputs500 = np.load(rec500)['test_out']
outputs1500 = np.load(rec1500)['test_out']
outputs2500 = np.load(rec2500)['test_out']
outputslist = [outputs500, outputs1500, outputs2500]
```

```python
#plot inputs
plt.figure(1)
for j in range(10):
    plt.subplot(1, 10, j+1)
    tmp_in = inputs[j*(-1)-1,0]
    plt.imshow(tmp_in.reshape(64,-1))
    plt.axis('off')

#plot outputs
for i, output in enumerate(outputslist):
    plt.figure(i+2)
    for j in range(10):
        plt.subplot(1, 10, j+1)
        tmp_out = output[j,0,:]
        plt.imshow(tmp_out.reshape(64,-1))
        plt.axis('off')
```
